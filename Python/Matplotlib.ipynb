```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple plot
x = np.linspace(-10,20)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# decision boundry
p1 = np.array([1,1,2,2,3,3,4,4])
p2 = np.array([4,5,4,5,1,2,1,2])

# Decision Boundry: wp+b=0 w=[2,-3] b=0
def boundryDecision(x):
    return (-2 - 0)/-3 * x

x = np.linspace(0,10)
y = boundryDecision(x)

plt.scatter(p1,p2, c="r", label="data")
plt.plot(x, y)
plt.xlabel("p1")
plt.ylabel("p2")
plt.legend()
plt.show()
```

```python
# scatter plot
x = np.array([1,1,2,2,3,3,4,4])
y = np.array([4,5,4,5,1,2,1,2])

fig, ax = plt.subplots()

ax.scatter(x, y, s = 10, color='#539caf', alpha = 0.75)
#ax.plot(x, y, lw=2, color='#539caf', alpha = 0.75)

plt.xlabel("p1")
plt.xlabel("p2")
plt.title("test scatter")
plt.show()
```

```python
# subplots
x = np.array([1,1,2,2,3,3,4,4])
y = np.array([4,5,4,5,1,2,1,2])

x1 = np.linspace(0,10)
y1 = np.cos(x1)
fig, ax = plt.subplots(2)
ax[0].scatter(x, y)
ax[1].plot(x1, y1)

plt.show()
```
