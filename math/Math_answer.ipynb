```python
# E.4:
# i.  Find the gradient vector of f (x;y) = x2+y2.
# ii. What are the gradient vectors at (1;2); (2;1)and(0;0)? Plot the function in 3D space and
#     check your answers?
def f(x,y):
    return (x**2 + y**2)
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)
fig = plt.figure(4)
ax = fig.gca(projection='3d')
ax.contour3D(X,Y,Z, 100, cmap="binary")
plt.show()
```

```python
# E.5:
# i.  Find the gradient vector of f (x;y) = 2xy+x2+y.
# ii. What are the gradient vectors at (1;1); (0;1)and(0;0)? Plot the function in 3D space and
#      check your answers?

def f(x,y):
    return (2*x*y + x**2 + y)
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)
fig = plt.figure(4)
ax = fig.gca(projection='3d')
ax.contour3D(X,Y,Z, 200)
plt.show()
```

```python
# E.7:
# Find the eigenvalues and eigenvectors of the given matrix by hand and check results by the
# computer (use Python to check your results).
# i)
a = np.array([[2, 0], [0, 5]])
w, v = np.linalg.eig(np.transpose(a))
print('Eigenvalues: ', w)
print('Eigenvectors: \n', v)
```

```python
# ii)
a = np.array([[5, 4], [1, 5]])
w, v = np.linalg.eig(np.transpose(a))
print('Eigenvalues: ', w)
print('Eigenvectors: \n', v)
```

```python
# iii)
a = np.array([[3, 3], [5, 1]])
w, v = np.linalg.eig(np.transpose(a))
print('Eigenvalues: ', w)
print('Eigenvectors: \n', v)
```
